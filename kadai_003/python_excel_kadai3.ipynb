```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# === Step3：Excelファイルの読み込み ===
df_2022 = pd.read_excel('2022_年間売上表.xlsx', sheet_name='Sheet1')
df_2023 = pd.read_excel('2023_年間売上表.xlsx', sheet_name='Sheet1')

# データの連結
df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# 商品・売上年ごとに集計
summary_df = (
    df_all
    .groupby(['商品', '売上年'], as_index=False)
    ['金額（千円）']
    .sum()
)

# === Step4：新規Excelファイルに書き出し ===
output_file = '売上集計表.xlsx'
summary_df.to_excel(output_file, index=False)

# openpyxlでヘッダーの書式変更
wb = load_workbook(output_file)
ws = wb.active

# 薄いグレーの塗りつぶし
header_fill = PatternFill(fill_type='solid', fgColor='F2F2F2')

# 1行目（ヘッダー）に適用
for cell in ws[1]:
    cell.fill = header_fill

# 保存
wb.save(output_file)
```
